```python
import numpy as np
import matplotlib.pyplot as plt

# Known standard deviation
sigma = 2

# Range of μ values
mu_values = np.linspace(-5, 5, 100)

# Observed data
x = np.array([1.5, 2.5, 3.0, 4.0, 4.5])  # Replace this with your actual data

# Initialize an empty array to store likelihood values
likelihood = np.zeros(len(mu_values))

# Calculate likelihood for each μ
for i, mu in enumerate(mu_values):
    likelihood[i] = np.prod(1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mu) / sigma)**2))

# Plot the likelihood
plt.plot(mu_values, likelihood)
plt.xlabel('μ')
plt.ylabel('Likelihood')
plt.title('Likelihood Function for μ')
plt.show()
```
