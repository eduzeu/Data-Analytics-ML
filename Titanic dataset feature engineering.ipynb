```python
# question 6: Write a method “data_split(data, ratio)” that shuffles the observations and splits--
 #the input data set by the input ratio between 0 and 1.--
 #Let the method return the split data sets

def data_split(data, ratio):
    '''based on the ratio, it will divide the dataset and shuffle it accordingly'''
 
    shuffled = data.sample(frac=1).reset_index(drop=True) 

    split = int(len(data) * ratio)
    data_set1 = shuffled[:split]
    data_set2 = shuffled[split:]

    
    
    return data_set1, data_set2

#example with ratio = 0.5
data1, data2 = data_split(data, 0.5)
nums = []
#this is only to test if there are repeated values, if there is an existing id, the loop will print it, if not, nothing happens
for passenger_id in data1['PassengerId']:
    if passenger_id  in nums:
        print("repeated id")
    else:
        nums.append(passenger_id)

#check if the split is correct 
if len(data1)+ len(data2) != len(data):
    print("datasets are not split correctly")
else:
    print("datasets split correctly")
    
#print datasets
data1, data2

```

```python
#Write a method “k_fold_CV(data, k)” that shuffles the observations and returns k many validation sets. 
#But the repetition of observations must be avoided

def k_fold_CV(data,k):
    '''shuffles observations and returns k sets'''

    shuffled = data.sample(frac=1).reset_index(drop=True)
    number_of_samples = len(data) / k
    datasets = []

    for i in range(k):
        start = int(i * number_of_samples)
        end = int((i + 1) * number_of_samples)

        if i == k - 1:
            # Handle any remaining observations in the last fold
            dataset_shuffled = shuffled[start:]
        else:
            dataset_shuffled = shuffled[start:end]

        start_split = shuffled[:start]
        end_split = shuffled[end:]

        train_set = pd.concat([start_split,end_split])
        datasets.append([train_set, dataset_shuffled])

    return datasets

#example with k = 2
data1 = k_fold_CV(data, 2)

sumsets = 0

#check if the datasets are correctly split k times
for i, (train_set, validation_set) in enumerate(data1):
    sumsets = sumsets + len(validation_set)
    print(f"Fold {i + 1} - Set Length: {len(validation_set)}")
    if sumsets == len(data):
        print("effectively divided the dataframe into k sets")

#check datasets
data1


```
